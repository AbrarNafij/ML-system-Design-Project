```python
import numpy as np
from scipy import stats

# 1. Automatically find the right columns even if names vary slightly
region_col = [c for c in sample_info.columns if 'Region' in c][0]
disease_col = [c for c in sample_info.columns if 'Disease' in c][0]
id_col = [c for c in sample_info.columns if 'Accession' in c][0]

print(f"Using columns: {id_col}, {region_col}, {disease_col}")

# 2. Filter for Entorhinal Cortex
region_mask = sample_info[region_col].str.contains('Entorhinal Cortex', na=False)
ec_metadata = sample_info[region_mask]

# 3. Get IDs for Normal and AD
# We use .str.strip() here too because the data inside the cells often has spaces
normal_ids = ec_metadata[ec_metadata[disease_col].str.contains('normal', case=False, na=False)][id_col].unique()
ad_ids = ec_metadata[ec_metadata[disease_col].str.contains('Alzheimer', case=False, na=False)][id_col].unique()

# 4. Extract from matrix (making sure the IDs exist in matrix columns)
normal_data = matrix_data[matrix_data.columns.intersection(normal_ids)]
ad_data = matrix_data[matrix_data.columns.intersection(ad_ids)]

print(f"Comparing {ad_data.shape[1]} AD samples vs {normal_data.shape[1]} Normal samples")

# 5. Statistical Test
t_stat, p_vals = stats.ttest_ind(ad_data, normal_data, axis=1, nan_policy='omit')

# 6. Results
results = pd.DataFrame({
    'Probe_ID': matrix_data.index,
    'P_Value': p_vals,
    'Log2FC': ad_data.mean(axis=1) - normal_data.mean(axis=1)
}).dropna()

top_genes = results.sort_values('P_Value').head(10)
print("\nTop 10 Most Significant Probes:")
print(top_genes)
```

```python
# Apply log2 transformation to the data
# We add 1 to avoid taking the log of zero (which is undefined)
matrix_data_log = np.log2(matrix_data + 1)

# Re-extract the groups with logged data
normal_logged = matrix_data_log[normal_data.columns]
ad_logged = matrix_data_log[ad_data.columns]

# Re-run the T-Test
t_stat, p_vals = stats.ttest_ind(ad_logged, normal_logged, axis=1, nan_policy='omit')

# Update results
results_logged = pd.DataFrame({
    'Probe_ID': matrix_data.index,
    'P_Value': p_vals,
    'Log2FC': ad_logged.mean(axis=1) - normal_logged.mean(axis=1)
}).dropna()

print("Top 10 Logged Results (Much more realistic FC values!):")
print(results_logged.sort_values('P_Value').head(10))
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. Identify columns again to ensure accuracy
region_col = [c for c in sample_info.columns if 'Region' in c][0]
disease_col = [c for c in sample_info.columns if 'Disease' in c][0]
id_col = [c for c in sample_info.columns if 'Accession' in c][0]

# 2. Get all unique regions, stripping non-breaking spaces (\xa0)
# This ensures we catch 'Hippocampus' even with hidden characters
sample_info[region_col] = sample_info[region_col].str.strip()
all_regions = sample_info[region_col].unique().tolist()

# 3. Create a clean Multi-Region Metadata table
multi_region_metadata = sample_info.copy()
multi_region_metadata['Target'] = multi_region_metadata[disease_col].apply(
    lambda x: 'AD' if 'Alzheimer' in str(x) or 'AD' in str(x) else 'Control'
)

# 4. Align Matrix and Transpose (Now including all 161 samples)
selected_ids = multi_region_metadata[id_col].tolist()
available_ids = [idx for idx in selected_ids if idx in matrix_data.columns]
full_matrix = matrix_data_log[available_ids].T

# 5. Feature Selection (Top 5000 Highly Variable Genes)
gene_variance = full_matrix.var(axis=0)
top_5000_genes = gene_variance.sort_values(ascending=False).head(5000).index
filtered_matrix = full_matrix[top_5000_genes]

# 6. Z-score Scaling for Deep Learning
scaler = StandardScaler()
scaled_data = scaler.fit_transform(filtered_matrix)
scaled_df = pd.DataFrame(scaled_data, index=filtered_matrix.index, columns=filtered_matrix.columns)

print(f"Phase 1 Reset Complete!")
print(f"Total Samples: {scaled_df.shape[0]}")
print(f"Total Genes: {scaled_df.shape[1]}")
print(f"Regions mapped: {all_regions}")
print(f"Sample breakdown:\n{multi_region_metadata[region_col].value_counts()}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# 1. Prepare Tensors
# We use the scaled_df from Phase 1
X_tensor = torch.FloatTensor(scaled_df.values)

# 2. Define the Contrastive Encoder Architecture
class ContrastiveEncoder(nn.Module):
    def __init__(self, input_dim, latent_dim=32):
        super(ContrastiveEncoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(1024, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim)
        )

    def forward(self, x):
        return self.encoder(x)

# 3. Model Setup
latent_dim = 32
model = ContrastiveEncoder(input_dim=5000, latent_dim=latent_dim)
optimizer = optim.Adam(model.parameters(), lr=0.0005)
criterion = nn.MSELoss() # Standard similarity proxy for representation learning

# 4. Self-Supervised Training Loop
epochs = 100
model.train()
print("Starting Self-Supervised Contrastive Training...")

for epoch in range(epochs):
    # Augmentation: Adding slight Gaussian noise (Simulating technical variance)
    noise = torch.randn_like(X_tensor) * 0.1
    augmented_X = X_tensor + noise

    optimizer.zero_grad()

    # Get representations (latents)
    z_original = model(X_tensor)
    z_augmented = model(augmented_X)

    # Minimize the distance between original and noisy version of same sample
    loss = criterion(z_original, z_augmented)

    loss.backward()
    optimizer.step()

    if (epoch + 1) % 20 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.6f}")

# 5. Extraction
model.eval()
with torch.no_grad():
    latent_representation = model(X_tensor).numpy()

print(f"\nPhase 2 Complete!")
print(f"Compressed 5,000 genes into {latent_dim} Latent Features.")
```

```python
# Create a Supervised Contrastive Loss proxy
# We will use the 'Target' labels to guide the encoder
y_tensor = torch.FloatTensor(y).view(-1, 1)

model = ContrastiveEncoder(input_dim=5000, latent_dim=32)
optimizer = optim.Adam(model.parameters(), lr=0.0005)

epochs = 150
for epoch in range(epochs):
    optimizer.zero_grad()

    # Forward pass
    latents = model(X_tensor)

    # Simpler Supervised Loss:
    # Try to make latents of the same class closer to each other
    # and different classes further apart using a distance matrix
    dist_matrix = torch.cdist(latents, latents)

    # Mask for same-class pairs
    labels_equal = (y_tensor == y_tensor.T).float()

    # Loss: Minimize distance for same class, maximize for different
    # (Simplified contrastive loss)
    loss = (labels_equal * dist_matrix).mean() - ((1 - labels_equal) * dist_matrix).mean()

    loss.backward()
    optimizer.step()

    if (epoch + 1) % 30 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# Re-extract features
model.eval()
with torch.no_grad():
    latent_representation = model(X_tensor).numpy()
```

Traditional AD research often focuses on the plaques themselves. However, your model prioritized CD44 and HLA-DQB1. This suggests that in the GSE5281 dataset, the immune system’s response to the disease is actually a more stable and "predictable" biomarker than the plaques themselves.

Biological Context: CD44 is a marker of reactive astrocytes. Its selection implies that "brain scarring" (gliosis) is a primary feature the AI uses to distinguish AD tissue across the Entorhinal Cortex and Superior Frontal Gyrus.

2. The Proteostasis and Energy Crisis (ATP6V1A and CDC37)
The selection of ATP6V1A (a lysosomal proton pump) and CDC37 (a chaperone) points to a "protein-handling crisis."

The Synergy: For a neuron to stay healthy, it must create energy and clear out "trash."

Mechanism: When ATP6V1A fails, lysosomes cannot become acidic enough to digest toxic proteins. When CDC37 fails, proteins aren't folded correctly in the first place. By selecting these, your AI identified that cellular waste management failure is a more robust indicator of AD than individual protein counts.

While APP is the most famous AD gene, SST (Somatostatin) is often a more precise marker for the specific loss of interneurons in the Hippocampus. Your model's preference for SST indicates it is sensitive to the loss of specific neuronal sub-populations, which is a key milestone in the progression from mild cognitive impairment to full Alzheimer’s.
